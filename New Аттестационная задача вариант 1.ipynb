```python
from math import gcd  # НОД


def plus(a, b):  # складывает a и b
    result = [str(int(a[0]) * int(b[2]) + int(a[2]) * int(b[0])), '/', str(int(a[2]) * int(b[2]))]
    return result


def minus(a, b):  # вычитает a из b
    result = [str(int(a[0]) * int(b[2]) - int(a[2]) * int(b[0])), '/', str(int(a[2]) * int(b[2]))]
    return result


def asterisk(a, b):  # умножает a на b
    result = [str(int(a[0]) * int(b[0])), '/', str(int(a[2]) * int(b[2]))]
    return result


def slash(a, b):  # делит a на b
    result = [str(int(a[0]) * int(b[2])), '/', str(int(a[2]) * int(b[0]))]
    return result


def operation(first, symbol, second):  # выбор операции
    if symbol == "+":
        return plus(first, second)
    elif symbol == "-":
        return minus(first, second)
    elif symbol == "*":
        return asterisk(first, second)
    elif symbol == "/":
        return slash(first, second)


def delete_spaces(s):  # удаляет все пробелы из списка
    return s.replace(" ", "")


def reformat_list(a):  # преобразует +- и -+ в -
    tmp = []
    i = 0
    while i < len(a) - 1:
        if a[i] == "+" and a[i + 1] == "-" or a[i] == "-" and a[i + 1] == "+":
            tmp.append("-")
            i += 2
        else:
            tmp.append(a[i])
            i += 1
    tmp.append(a[len(a) - 1])
    return tmp


def merge_numbers(a):  # соединяет цифры в числа в списке
    tmp = ""
    tmp_l = []
    for i in a:
        if i.isdigit():
            tmp += i
        else:
            if tmp != "":
                tmp_l.append(tmp)
            tmp = ""
            tmp_l.append(i)
    if tmp != "":
        tmp_l.append(tmp)
    return tmp_l


def is_string_valid(s):  # проверка начала и конца строки
    incorrect = ["+", "*", "/", "^"]
    flag = False
    if s[0] == ")" or s[0] in incorrect:
        print("Error: Скорее всего, строка введена не полностью")
        return 0
    if s[len(s) - 1] == "(" or s[len(s) - 1] in incorrect or s[len(s) - 1] == "-":
        print("Error: Скорее всего, строка введена не полностью")
        return 0
    for i in s:
        if i in incorrect or i == "-":
            flag = True
    if flag:
        return 1
    else:
        print("Error: Скорее всего, строка введена не полностью")
        return 0


def is_symbols_valid(s):  # проверка символов в строке
    correct_symbols = ["1", "2", "3", "4", "5", "6", "7", "8", "9",
                       "0", "+", "-", "*", "/", "^", "(", ")", " "]
    for i in s:
        if i not in correct_symbols:
            print("Error: Введён некорректный символ в строку")
            return 0
    return 1


def parentheses_check(s):  # проверка порядка и равенства количества скобок
    in_count = 0
    out_count = 0
    for i in s:
        if i == "(":
            in_count += 1
        if i == ")":
            if in_count - out_count >= 1:
                out_count += 1
            else:
                print("Error: Где-то не хватает скобки")
                return 0
    if in_count == out_count:
        return 1
    else:
        print("Error: Где-то не хватает скобки")
        return 0


def fractions_count(s):  # количество смешанных дробей в строке
    count = 0
    for i in range(len(s) - 1):
        if s[i + 1] == "(" and s[i][0].isdigit():
            count += 1
    return count


def parentheses_count(s):  # количество скобок в строке
    count = 0
    for i in s:
        if i == "(":
            count += 1
    return count


def calculate_fractions(a):  # переводит смешанные дроби в неправильные
    tmp = []
    start = None
    number = None
    for i in range(len(a) - 1):
        if a[i + 1] == "(" and a[i][0].isdigit():
            number = i
            start = i + 2
    i = start
    if start is not None:
        i = 0
        while i < len(a):
            if i != start - 1:
                tmp.append(a[i])
                i += 1
            else:
                tmp.pop()
                tmp.append([str(int(a[number][0]) * int(a[start][2]) + int(a[start][0])), "/", int(a[start][2])])
                i = start + 2
    return tmp


def calculate_parentheses(a):  # считает значение в скобочке
    tmp = []
    start = None
    end = None
    for i in range(len(a)):
        if a[i] == ")":
            end = i
            break
    for i in range(len(a)):
        if a[i] == "(" and i < end:
            start = i + 1
    if start is not None and end is not None:
        i = 0
        while i < len(a):
            if i != start - 1:
                tmp.append(a[i])
                i += 1
            else:
                tmp.append(*calculation(a[start:end]))
                i = end + 1
    return tmp


def string_to_list(s):  # Делает список, с которым удобнее работать
    lst = []
    i = 0
    while i < len(s):
        if s[i] == "/" and s[i - 1][0].isdigit() and s[i + 1][0].isdigit():
            lst.pop()
            tmp = list()
            tmp.append(s[i - 1])
            tmp.append("/")
            tmp.append(s[i+1])
            lst.append(tmp)
            i += 2
        else:
            lst.append(s[i])
            i += 1
    s = lst
    lst = []
    for i in s:
        if type(i) != list and i[0].isdigit():
            tmp = list()
            tmp.append(i)
            tmp.append("/")
            tmp.append("1")
            lst.append(tmp)
        elif i != " ":
            lst.append(i)
    return lst


def calculation(a):  # считает значение в строке
    first_prior = ["*", "/"]
    second_prior = ["+", "-"]
    while len(a) != 1:
        first_flag = False
        second_flag = False
        i = 0
        tmp = []
        while i < len(a):
            if a[i] in first_prior and not first_flag:
                tmp.pop()
                result = operation(a[i - 1], a[i], a[i + 1])
                tmp.append(result)
                i += 2
                first_flag = True
            else:
                tmp.append(a[i])
                i += 1
        a = tmp
        i = 0
        tmp = []
        if not first_flag:
            while i < len(a):
                if a[i] in second_prior and not second_flag:
                    tmp.pop()
                    result = operation(a[i - 1], a[i], a[i + 1])
                    tmp.append(result)
                    i += 2
                    second_flag = True
                else:
                    tmp.append(a[i])
                    i += 1
            a = tmp
    return a


def formatting(a):  # вывод ответа
    answer = ""
    if int(a[0]) % int(a[2]) == 0:
        answer += str(int(int(a[0]) / int(a[2])))
    elif int(a[0]) > 0:
        tmp_gcd = gcd(int(int(a[0]) % int(a[2])), int(a[2]))
        if int(int(a[0]) // int(a[2])) != 0:
            answer += str(int(int(a[0]) // int(a[2])))
            answer += "("
        answer += str(int(int(a[0]) % int(a[2]) / tmp_gcd))
        answer += "/"
        answer += str(int(int(a[2]) / tmp_gcd))
        if int(int(a[0]) // int(a[2])) != 0:
            answer += ")"
    elif int(a[0]) < 0:
        tmp_gcd = gcd(-int(int(a[0]) % -int(a[2])), -int(a[2]))
        if -int(int(a[0]) // -int(a[2])) != 0:
            answer += str(-int(int(a[0]) // -int(a[2])))
            answer += "("
        answer += "-"
        answer += str(-int(int(a[0]) % -int(a[2]) / tmp_gcd))
        answer += "/"
        answer += str(int(int(a[2]) / tmp_gcd))
        if -int(int(a[0]) // -int(a[2])) != 0:
            answer += ")"

    return answer


def test(a):
    answer = ""
    if int(a[0]) % int(a[2]) == 0:
        answer += str(int(int(a[0]) / int(a[2])))


def save_result(key, results, result):
    if key not in results.keys():
        results[key] = result
        print("save success")
    else:
        print("remove the old value before saving new")
    return results


def del_result(key, results):
    if key in results.keys():
        results.pop(key)
        print("delete success")
    else:
        print("this variable doesn't exists")
    return results


def find_results(s, results):
    for key in results.keys():
        if key in s:
            s = s.replace(key, results[key])
    return s


def main():  # главная функция
    result = None
    result_double = None
    results = {}
    string = input("Input: ")
    while string != "exit":
        if "save" in string:  # еще не сделано
            results = save_result(string[5:], results, result)
        elif "del" in string:  # еще не сделано
            results = del_result(string[4:], results)
        elif string == "to double":
            if result is not None:
                print("Output:", int(result_double[0][0]) / int(result_double[0][2]))
        else:
            string = find_results(string, results)
            string = reformat_list(string)
            if is_string_valid(string) and is_symbols_valid(string) and parentheses_check(string):
                string = merge_numbers(string)
                string = string_to_list(string)
                parentheses = parentheses_count(string)
                fractions = fractions_count(string)
                for i in range(fractions):
                    string = calculate_fractions(string)
                for i in range(parentheses - fractions):
                    string = calculate_parentheses(string)
                result = calculation(string)
                result_double = result
                result = formatting(*result)
                print("Output:", result)
        string = input("Input: ")


main()
```
